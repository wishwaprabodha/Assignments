```python
#unsupervised learning 
#clustering,Association Rule Discovery.....

import pandas as pd
print(pd.__version__)
```

```python
original_data=pd.read_csv("./winedata.csv")

original_data.head()
```

```python
training_data=original_data.copy()
del training_data['Type']
training_data.head()
```

```python
from sklearn.cluster import KMeans
kmeans= KMeans(n_clusters=3,random_state=0)
result=kmeans.fit(training_data)
result.labels_
```

```python
result.inertia_
```

```python
original_data=pd.read_csv("./winedata.csv")
training_data=original_data.copy()
del training_data['Type']
training_data.head()

sse={}

for k in range(1,20) :
    kmeans=KMeans(n_clusters=k,random_state=0)
    results=kmeans.fit(training_data)
    sse[k]=results.inertia_
    print(sse[k],'\n')
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(list(sse.keys()),list(sse.values()))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.xticks(list(sse.keys()))

print("the best num of clusters are at the knee bend because minim deviation from one SSE another SSE closer cluster ")
```
